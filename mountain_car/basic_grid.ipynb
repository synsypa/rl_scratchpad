```python
import gym
import numpy as np
import matplotlib.pyplot as plt
import random
from mountain_car import MountainCar
%matplotlib inline
```

```python
env = gym.make('MountainCar-v0')
env.reset()
episodes = 5000
alpha = .1
gamma = 1
epsilon = .75
decay = epsilon/episodes
sarsa_car = MountainCar(env, 'sarsa', alpha, gamma, epsilon)
expected_car = MountainCar(env, 'expected_sarsa', alpha, gamma, epsilon)
q_car = MountainCar(env, 'q', alpha, gamma, epsilon)
```

```python
for i in range(episodes):
        sarsa_car.run_episode(decay)
        expected_car.run_episode(decay)
        q_car.run_episode(decay)
```

```python
plt.plot(100*(np.arange(len(sarsa_car.reward_list)) + 1), sarsa_car.reward_list, 'B')
plt.plot(100*(np.arange(len(expected_car.reward_list)) + 1), expected_car.reward_list, 'R')
plt.plot(100*(np.arange(len(q_car.reward_list)) + 1), q_car.reward_list, 'G')
plt.xlabel('Episodes')
plt.ylabel('Smoothed Reward')
plt.title('Smoothed Reward vs Episodes')
plt.show()
```
